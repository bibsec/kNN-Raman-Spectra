```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.figure(figsize=(9, 6))

def spectral(x, k):
    y = (x**2) *k*0.25
    return y
g = np.linspace(0, 1, 1000000)
#print(g)
v = spectral(g, 3500)
#print(v)
plt.plot(g, v)
plt.ylim(0, 10)
plt.xlim(0,0.107)
plt.ylabel(r'$\Delta \sigma (cm^{-1})$', size = 14)
plt.xlabel(r'$D_d\,/\,\mathrm{D_s}$', size = 14)
plt.title('Spectral Resolution as a function of ratio of Diameters', size = 16)
```

```python
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi
from scipy.stats import chisquare
%matplotlib inline


Starkeyite = "https://rruff.info/repository/sample_child_record_raman/by_minerals/Starkeyite__R060321__Raman__780__0__unoriented__Raman_Data_Processed__34110.txt"
scdat = np.recfromtxt(Starkeyite, names=['Raman_shift', 'intensity'], skip_header=0,  delimiter=',')
Raman = scdat.Raman_shift
intensity = scdat.intensity
f_exp = np.sqrt(intensity**2 + Raman**2)
print(len(f_exp))
#print(f_exp)
a = [Raman, intensity]
exp = np.array([Raman, intensity]).T
#exp[exp == 0] = -1
#print(exp)
chi2 = chisquare(exp, ddof=2477)
#print(exp.shape)
#print(chi)



for i in range(0,16):
    plt.figure(figsize=(12, 10))
    #plt.plot(Raman, intensity, label='Original')
    plt.plot(a[0], a[1], label='Original')
    d1 = gaussian_filter(a[0], sigma=i)
    d2 = gaussian_filter(a[1], sigma=i)
    obs = np.array([d1, d2]).T
    f_obs = np.sqrt(d1**2 +d2**2)
    #print(f_obs)
    #obs[obs == 0] = 1
    chi1 = chisquare(f_obs, f_exp, ddof= 0)
    print(chi1)
    #print(d[:,0])
    #plt.plot(d[:,0], d[:,1] , '--', label='Gaussian convolution')
    #print(a)
    #print(d)
    plt.plot(d1, d2 , 'k.', label='Gaussian convolution')
    plt.ylabel('Intensity', size = 14)
    plt.xlabel('Raman shift', size = 14)
    plt.title(('Starkeyite-Raman spectrum with sigma equal to %i' %(i)), size = 16)
    #plt.xlim(1050, 1200)
    #plt.ylim(0, 80)
    plt.legend()
    plt.grid()
    plt.show();
    
```
